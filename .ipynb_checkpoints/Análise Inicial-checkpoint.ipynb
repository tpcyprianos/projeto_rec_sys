### Considerando parte da base KT1
O dataset EdNet-KT1 possui **268.683 usuários/arquivos**, neste caso, serão considerados os **1000 usuários**.

```python
import pandas as pd
import glob
import os

# Defina o caminho da pasta onde estão os arquivos CSV
caminho_pasta = r'D:\Doutorado\USP\SCC5966 - Recomendaçao\Projeto\dataset\EdNet-KT1\KT1'
caminho_arquivos = os.path.join(caminho_pasta, '*.csv')

# Use glob para encontrar todos os arquivos CSV na pasta
arquivos_csv = glob.glob(caminho_arquivos)

# Ordenar os arquivos pelo tamanho (do maior para o menor)
#arquivos_csv = sorted(arquivos_csv, key=os.path.getsize, reverse=True)
##### Esta estratégia deixou a execução muito demorada

# Limite para os 1000 primeiros arquivos
arquivos_csv = arquivos_csv[:1000]

# Crie uma lista para armazenar os DataFrames
dataframes = []

# Carregue cada arquivo CSV e adicione o DataFrame à lista
for arquivo in arquivos_csv:
    # Lê o arquivo CSV
    df = pd.read_csv(arquivo)
    
    # Adiciona uma coluna com o nome do arquivo (sem o caminho completo)
    df['codigo_arquivo'] = os.path.basename(arquivo)
    
    # Adiciona o DataFrame à lista
    dataframes.append(df)

# Opcional: Combine todos os DataFrames em um único DataFrame (caso tenham o mesmo formato)
kt1 = pd.concat(dataframes, ignore_index=True)

# Exibe as primeiras linhas do DataFrame combinado para verificar
print(kt1.head())
```

```python
dimensao_kt1 = kt1.shape 
print("Dimensão do dataset KT1 é ", dimensao_kt1[0], " linhas e ", dimensao_kt1[1], " colunas.")
```

```python
# Criar uma nova coluna com o identificador do usuário, removendo o ".csv"
kt1['usuario'] = kt1['codigo_arquivo'].str.replace('.csv', '', regex=False)
kt1.head()
```

```python
kt1_usuarios = kt1['usuario'].unique()
kt1_usuarios
```

### Carregando os dados correspondentes dos usuários no dataset KT2

```python
import pandas as pd
import os

# Crie uma lista para armazenar os DataFrames
dataframes_kt2= []

# Caminho da pasta onde estão os arquivos
caminho_pasta = r'D:\Doutorado\USP\SCC5966 - Recomendaçao\Projeto\dataset\EdNet-KT2\KT2'

# Iterar sobre os nomes na coluna 'usuario'
for usuario in kt1_usuarios:
    # Monta o nome completo do arquivo (adiciona o sufixo ".csv")
    nome_arquivo = f"{usuario}.csv"
    caminho_arquivo = os.path.join(caminho_pasta, nome_arquivo)
    
    # Verifica se o arquivo existe na pasta
    if os.path.exists(caminho_arquivo):
        # Lê o arquivo e adiciona a coluna com o nome do arquivo
        temp_df = pd.read_csv(caminho_arquivo)
        temp_df['codigo_arquivo'] = nome_arquivo  # Adiciona a coluna com o nome do arquivo
        # Adiciona uma coluna com o nome do arquivo (sem o caminho completo)
        # Adiciona o DataFrame à lista
        dataframes_kt2.append(temp_df)
    #else:
        #print("Arquivo ", usuario, " não existe!")
kt2 = pd.concat(dataframes_kt2, ignore_index=True)
```
