```python
import pandas as pd
import glob
import os

# Defina o caminho da pasta onde estão os arquivos CSV
caminho_pasta = r'D:\Doutorado\USP\SCC5966 - Recomendaçao\Projeto\dataset\EdNet-KT1\KT1'
caminho_arquivos = os.path.join(caminho_pasta, '*.csv')

# Use glob para encontrar todos os arquivos CSV na pasta
arquivos_csv = glob.glob(caminho_arquivos)

# Crie uma lista para armazenar os DataFrames
dataframes = []

# Carregue cada arquivo CSV e adicione o DataFrame à lista
for arquivo in arquivos_csv:
    df = pd.read_csv(arquivo)
    dataframes.append(df)

# Opcional: Combine todos os DataFrames em um único DataFrame (caso tenham o mesmo formato)
df_combinado = pd.concat(dataframes, ignore_index=True)

# Exibe as primeiras linhas do DataFrame combinado para verificar
print(df_combinado.head())
```

## Considerando parte da base KT1
O dataset EdNet-KT1 possui **268.683 usuários/arquivos**, neste caso, serão considerados os **10.000 usuários** com mais interações

```python
import pandas as pd
import glob
import os

# Defina o caminho da pasta onde estão os arquivos CSV
caminho_pasta = r'D:\Doutorado\USP\SCC5966 - Recomendaçao\Projeto\dataset\EdNet-KT1\KT1'
caminho_arquivos = os.path.join(caminho_pasta, '*.csv')

# Use glob para encontrar todos os arquivos CSV na pasta
arquivos_csv = glob.glob(caminho_arquivos)

# Ordenar os arquivos pelo tamanho (do maior para o menor)
arquivos_csv = sorted(arquivos_csv, key=os.path.getsize, reverse=True)

# Limite para os 10000 maiores arquivos
arquivos_csv = arquivos_csv[:10000]

# Crie uma lista para armazenar os DataFrames
dataframes = []

# Carregue cada arquivo CSV e adicione o DataFrame à lista
for arquivo in arquivos_csv:
    # Lê o arquivo CSV
    df = pd.read_csv(arquivo)
    
    # Adiciona uma coluna com o nome do arquivo (sem o caminho completo)
    df['codigo_arquivo'] = os.path.basename(arquivo)
    
    # Adiciona o DataFrame à lista
    dataframes.append(df)

# Opcional: Combine todos os DataFrames em um único DataFrame (caso tenham o mesmo formato)
kt1 = pd.concat(dataframes, ignore_index=True)

# Exibe as primeiras linhas do DataFrame combinado para verificar
print(kt1.head())
```

```python
kt1.to_csv('kt1.txt', sep='\t', index=False, header=False)
```
